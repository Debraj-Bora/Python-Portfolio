```python
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
```

## Importing the file for analysis

```python
warnings.filterwarnings('ignore')
df = pd.read_csv("ultra_marathon.csv")
df.head()
```

## Data Cleaming and Preparation

```python
df.shape
```

```python
df.dtypes
```

- Checking events in  USA (50k or 50 Miles) and the year 2020

```python
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
df[ (df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
df[ (df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df_usa = df[ (df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df_usa.head()
```

```python
df_usa.shape
```

- Remove (USA) from event name

```python
df_usa['Event name'] = df_usa['Event name'].str.split('(').str.get(0)
```

```python
df_usa.head()
```

- Calculating the age of athletes

```python
df_usa['athlete_age'] = 2020 - df_usa['Athlete year of birth']
```

- remove h from athlete performance

```python
df_usa['Athlete performance'] = df_usa['Athlete performance'].str.split(' ').str.get(0)
```

```python
df_usa.head()
```

## Dropping columns not relevant to our analysis
- Athlete Club
- Athlete Country
- Athlete year of birth
- Athlete Age Category

```python
df_usa = df_usa.drop(['Athlete club', 'Athlete year of birth', 'Athlete age category', 'Athlete country' ], axis = 1)
```

```python
df_usa.head()
```

- clean up null values

```python
df_usa.isna().sum()
```

```python
df_usa[df_usa['athlete_age'].isna()]
```

```python
df_usa = df_usa.dropna()
```

```python
df_usa.shape
```

- Check for duplicates

```python
df_usa[df_usa.duplicated() == True]
```

- reset index

```python
df_usa.reset_index(drop = True)
```

- fixing the data types of columns

```python
df_usa.dtypes
```

```python
df_usa['athlete_age'] = df_usa['athlete_age'].astype(int)
```

```python
df_usa['Athlete average speed'] = df_usa['Athlete average speed'].astype(float)
```

```python
df_usa.dtypes
```

- rename columns

```python
Year of event                  int64
Event dates                   object
Event name                    object
Event distance/length         object
Event number of finishers      int64
Athlete performance           object
Athlete gender                object
Athlete average speed         object
Athlete ID                     int64
athlete_age                  float64
```

```python
df_usa = df_usa.rename(columns = {
    'Year of event' : 'year',
    'Event dates' : 'race_date',
    'Event name' : 'race_name',
    'Event distance/length' : 'race_length',
    'Event number of finishers' : 'race_number_of_finishers',
    'Athlete performance' : 'athlete_performance',
    'Athlete gender' : 'gender',
    'Athlete average speed' : 'athlete_average_speed',
    'Athlete ID' : 'athlete_id'
})

df_usa.head()
```

- dropping the year column as all the races are in 2020

```python
df_usa = df_usa.drop('year', axis =1 )
```

```python
df_final = df_usa[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
df_final.head()
```

## Visualizing the data

```python
# race count by gender
```

```python
sns.histplot(data = df_final, x = 'race_length', hue = 'gender')
plt.show()
```

```python
# distribution plot of average speed
```

```python
sns.displot(df_final[df_final['race_length'] == '50mi']['athlete_average_speed'])
plt.show()
```

```python
sns.violinplot(data = df_final, x = 'race_length', y = 'athlete_average_speed', hue = 'gender', split = True, inner = 'quart', linewidth = 1)
plt.show()
```

```python
sns.lmplot(data = df_final, x = 'athlete_age', y = 'athlete_average_speed', hue = 'gender')
plt.show()
```

## Insights

- difference in speed for the 50k, 50mi male to female

```python
df_final.groupby(['race_length', 'gender'])['athlete_average_speed'].mean()
```

- which age groups are the best in the 50m Race (20+ races min)

```python
df_final.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

- which age groups are slowest in the 50m Race (20+ races min)

```python
df_final.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>19').head(15)
```

## impact of season on runners
- spring 3 - 5
- summer 6 - 8
- fall 9 - 11
- winter 12 -2

```python
df_final['race_month'] = df_final['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df_final['race_season'] = df_final['race_month'].apply(lambda x : 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df_final.head()
```

```python
df_final.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
